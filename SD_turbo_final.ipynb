```python
import os, io, csv, requests, wget
import pandas as pd
from PIL import Image

# Paths
TARGET = "/content/diffusers/examples/text_to_image/pokemon_dataset"
IMAGES = os.path.join(TARGET, "images")
CAPTIONS = os.path.join(TARGET, "captions.csv")
PARQUET_PATH = "/content/train.parquet"
PARQUET_URL = "https://huggingface.co/datasets/reach-vb/pokemon-blip-captions/resolve/main/data/train-00000-of-00001-566cc9b19d7203f8.parquet"

# Create directories
os.makedirs(IMAGES, exist_ok=True)

# Download Parquet file
if not os.path.exists(PARQUET_PATH):
    wget.download(PARQUET_URL, PARQUET_PATH)

# Load dataset
df = pd.read_parquet(PARQUET_PATH)

# Save images and captions
with open(CAPTIONS, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["filename", "caption"])
    for idx, row in df.iterrows():
        img_field = row["image"]
        img_bytes = img_field.get("bytes") if isinstance(img_field, dict) else img_field
        if hasattr(img_bytes, "to_pybytes"):
            img_bytes = img_bytes.to_pybytes()
        img = Image.open(io.BytesIO(img_bytes)).convert("RGB")
        fn = f"pokemon_{idx:05d}.png"
        img.save(os.path.join(IMAGES, fn))
        writer.writerow([fn, row["text"]])

print("✅ Dataset successfully saved to", TARGET)
```

#Building the Dataset

```python
import os
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

class Pokemon100Dataset(Dataset):
    def __init__(self, root_dir, captions_file, tokenizer, image_size=256):
        self.root = root_dir
        self.df = pd.read_csv(captions_file)
        self.tokenizer = tokenizer
        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize([0.5]*3, [0.5]*3)
        ])

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        # Load & preprocess image
        img_path = os.path.join(self.root, "images", row["image_filename"])
        img = Image.open(img_path).convert("RGB")
        pixel_values = self.transform(img)
        # Tokenize caption
        inputs = self.tokenizer(
            row["caption"],
            padding="max_length",
            truncation=True,
            max_length=64,
            return_tensors="pt",
        )
        input_ids = inputs.input_ids[0]
        attention_mask = inputs.attention_mask[0]
        return {"pixel_values": pixel_values,
                "input_ids": input_ids,
                "attention_mask": attention_mask}

# Usage:
from transformers import CLIPTokenizer
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-large-patch14")

dataset = Pokemon100Dataset(
    root_dir="/content/first_100",
    captions_file="/content/first_100/captions.csv",
    tokenizer=tokenizer,
    image_size=256,
)
loader = DataLoader(dataset, batch_size=1, shuffle=True, num_workers=0)
# Quick check:
batch = next(iter(loader))
print(batch["pixel_values"].shape, batch["input_ids"].shape)
# → torch.Size([1,3,256,256]) torch.Size([1,64])
```

```python
import os
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from transformers import CLIPTokenizer

# 1) Define the Dataset
class Pokemon100Dataset(Dataset):
    def __init__(self, root_dir, captions_file, tokenizer, image_size=256):
        self.root = root_dir
        self.df = pd.read_csv(captions_file)
        self.tokenizer = tokenizer
        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize([0.5]*3, [0.5]*3),
        ])

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(os.path.join(self.root, "images", row["image_filename"])).convert("RGB")
        pixel_values = self.transform(img)
        tokens = self.tokenizer(
            row["caption"],
            padding="max_length",
            truncation=True,
            max_length=64,
            return_tensors="pt"
        )
        return {
            "pixel_values": pixel_values,             # [3,256,256]
            "input_ids": tokens.input_ids[0],         # [64]
            "attention_mask": tokens.attention_mask[0]
        }

# 2) Instantiate tokenizer & dataset
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-large-patch14")
dataset = Pokemon100Dataset(
    root_dir="/content/first_100",
    captions_file="/content/first_100/captions.csv",
    tokenizer=tokenizer,
)

# 3) Create the DataLoader named `loader`
loader = DataLoader(dataset, batch_size=1, shuffle=True, num_workers=0)

# 4) Quick sanity check
batch = next(iter(loader))
print(batch["pixel_values"].shape, batch["input_ids"].shape)
# Should print: torch.Size([1,3,256,256]) torch.Size([1,64])
```

#ABCD

```python
from diffusers import StableDiffusionPipeline
import torch

# Load pipeline with balanced CPU placement
pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/sd-turbo",
    torch_dtype=torch.float32,
    safety_checker=None,
    device_map="balanced",
)
pipe.enable_attention_slicing()

# Freeze UNet & VAE; leave only the text encoder trainable
for p in pipe.unet.parameters():
    p.requires_grad = False
for p in pipe.vae.parameters():
    p.requires_grad = False
for p in pipe.text_encoder.parameters():
    p.requires_grad = True

print("✅ Model prepared. Text encoder trainable:",
      sum(p.requires_grad for p in pipe.text_encoder.parameters()),
      "params.")
```

```python
from torch.optim import AdamW

optimizer = AdamW(pipe.text_encoder.parameters(), lr=5e-5)
```

```python
import torch.nn.functional as F
from tqdm import tqdm

num_epochs = 3

pipe.text_encoder.train()
for epoch in range(num_epochs):
    loop = tqdm(loader, desc=f"Epoch {epoch+1}/{num_epochs}")
    for batch in loop:
        optimizer.zero_grad()

        # 1) Encode images → latents
        pixel_values = batch["pixel_values"]             # [1,3,256,256]
        latents = pipe.vae.encode(pixel_values).latent_dist.sample()

        # 2) Pick a single timestep
        t = torch.tensor([1], dtype=torch.long)

        # 3) Add noise
        noise = torch.randn_like(latents)

        # 4) Get text embeddings (trainable)
        encoder_states = pipe.text_encoder(
            batch["input_ids"], attention_mask=batch["attention_mask"]
        ).last_hidden_state

        # 5) Predict noise via frozen UNet
        noise_pred = pipe.unet(
            latents, t, encoder_hidden_states=encoder_states
        ).sample

        # 6) Compute loss & backprop only into text encoder
        loss = F.mse_loss(noise_pred, noise)
        loss.backward()
        optimizer.step()

        loop.set_postfix(loss=loss.item())
```
